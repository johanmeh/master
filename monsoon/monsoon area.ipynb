```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(15,10))
m = Basemap(width=8000000,height=7000000,projection='lcc',
            resolution=None,lat_1=-10.,lat_2=40,lat_0=10.,lon_0=80.)
m.shadedrelief()
parallels = np.arange(-10,81,10.)
# labels = [left,right,top,bottom]
m.drawparallels(parallels,labels=[False,True,False,False],fontsize=18)
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[False,False,False,True], fontsize=18)
plt.title('The Asian monsoon area', fontsize=18)
#plt.savefig('figs/monsoon_area_new.png', dpi=100)
```
