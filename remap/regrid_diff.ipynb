```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds = xr.open_dataset('regrid_era5_remapnn.nc')
ds2 = xr.open_dataset('gridfile_test.nc')
```

```python
ds_nn = ds.sel(lat=slice(-20,20))
ds_bil = ds2.sel(lat=slice(-20,20))
```

```python
ds_nn.ctpt[:,:,:].mean(axis=0).plot()
```

```python
ds_bil.ctpt[:,:,:].mean(axis=0).plot()
```

```python
diff = ds_nn.ctpt[:,:,:].mean(axis=0) - ds_bil.ctpt[:,:,:].mean(axis=0)
```

```python
diff.plot()
```

```python
diff_lrt = ds_nn.tpt[:,:,:].mean(axis=0) - ds_bil.tpt[:,:,:].mean(axis=0)
```

```python
diff_lrt.plot()
```
